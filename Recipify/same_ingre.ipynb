```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
input_df = pd.read_csv('ingre_sample.csv', encoding="cp949")
basic_df = pd.read_csv('recipe_basic2.csv', encoding="cp949")
```

```python
input_value = input_df["ingre"][0].split(", ")
ingre_data = basic_df["INGRE_NM"]
#사용자 입력 데이터
input_value2 = ['감자', '양파', '대파', '돼지고기', '소금', '설탕', '당근', '청양고추', '밥']
```

```python
result_dict = {"index" : [],
               "ingre" : [],
               "count_ingre" : []}

for i, ingre_v in enumerate(ingre_data):        
    same_list = []

    for input_v in input_value2:
        for same in ingre_v.split(", "):
            if input_v == same:
                same_list.append(input_v)
                break

    result_dict["index"].append(i)
    result_dict["ingre"].append(same_list)
    result_dict["count_ingre"].append(len(same_list))
```

```python
intersection = pd.DataFrame(result_dict).sort_values(by=['count_ingre'], axis=0, ascending=False)
```

```python
intersection = intersection[intersection.count_ingre > len(input_value2)/3]
```

```python
import pandas as pd

tf_idf_result = pd.read_csv("similar_recipes_top_10_tf_idf.csv")
img_url = pd.read_csv("recipe_dataset/db/recipe_basic.csv", encoding="cp949")
```

```python
tf_idf_result = tf_idf_result[["Recipe_ID", "Similar_Recipe_ID", "TF-IDF Value"]]
img_url = img_url[["RECIPE_NM_KO", "IMG_URL"]]

drop_list = list(tf_idf_result[tf_idf_result["Recipe_ID"] == tf_idf_result["Similar_Recipe_ID"]].index)

tf_idf_result = tf_idf_result.drop(drop_list).reset_index(drop=True)
```

```python
tf_idf_result_dict = {"Recipe_ID":[], 
                      "Similar_Recipe_ID":[], 
                      "TF-IDF Value":[], 
                      "IMG_URL":[]}

for i_menu in img_url["RECIPE_NM_KO"]:
    temp = tf_idf_result[tf_idf_result["Recipe_ID"] == i_menu].reset_index(drop=True).iloc[:5, :]
    
    for t_i, t_menu in enumerate(temp["Recipe_ID"]):
        recipe_id = t_menu
        similar_r_id = temp["Similar_Recipe_ID"][t_i]
        tf_idf_v = temp["TF-IDF Value"][t_i]
        map_img_url = img_url[img_url["RECIPE_NM_KO"] == similar_r_id]["IMG_URL"]
        map_img_url_index = map_img_url.index
        
        
        tf_idf_result_dict["Recipe_ID"].append(recipe_id)
        tf_idf_result_dict["Similar_Recipe_ID"].append(similar_r_id)
        tf_idf_result_dict["TF-IDF Value"].append(tf_idf_v)
        tf_idf_result_dict["IMG_URL"].append(map_img_url[map_img_url_index[0]])
```

```python
total_df = pd.DataFrame(tf_idf_result_dict)
total_df
```

```python
total_df.to_csv("TF_IDF_data.csv", index = False, encoding="utf-8-sig")
```
